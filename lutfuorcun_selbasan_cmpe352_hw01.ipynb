```python
from thinkdsp import read_wave

#read_wave takes .wav as argument and returns normalized Wave object
waveClear = read_wave('119200063_lutfuorcun_selbasan_cmpe352_hw01_clear.wav')
waveNoise = read_wave('119200063_lutfuorcun_selbasan_cmpe352_hw01_noise.wav')
```

```python
# Audio of clean sound
waveClear.make_audio()
```

```python
# Audio of noisy sound
waveNoise.make_audio()
```

### Graph of clear wave

```python
waveClear.plot()
```

### Graph of noisy wave

```python
waveNoise.plot()
```

### Graph of clear sound between the seconds 0.5 and 2.0

```python
# Segment extracts a part of the wave and returns it
segmentClear = waveClear.segment(start=0.5, duration= 1.5)
segmentClear.plot()
```

### Graph of noisy sound between the seconds 0.5 and 2.0

```python
# Segment extracts a part of the wave and returns it
segmentNoise = waveNoise.segment(start=0.5, duration= 1.5)
segmentNoise.plot()
```

```python
# Creating the spectrums from segments
spectrumClear = segmentClear.make_spectrum()
spectrumNoise = segmentNoise.make_spectrum()
```

```python
spectrumClear.plot(8000)
```

```python
spectrumNoise.plot(high= 5000)
```

```python
# Filtering the spectrum which attenuate frequencies above the cutoff(2000)
spectrumClear.low_pass(2000)
filteredClear = spectrumClear.make_wave()
filteredClear.make_audio()
```

```python
# after low_pass(2000)
spectrumClear.plot(8000)
```

```python
# Reassigns the old spectrum, filters it and converts it to audio
spectrumClear = segmentClear.make_spectrum()
spectrumClear.low_pass(10000)
filteredClear = spectrumClear.make_wave()
filteredClear.make_audio()
```

```python
# after low_pass(10000)
spectrumClear.plot(8000)
```

```python
# Filtering the spectrum which attenuate frequencies below the cutoff(1000)
spectrumNoise.high_pass(1000)
filteredNoise = spectrumNoise.make_wave()
filteredNoise.make_audio()
```

```python
# after high_pass(1000)
spectrumNoise.plot(high= 10000)
```

```python
# Reassigns the old spectrum, filters it and converts it to audio
spectrumNoise = segmentNoise.make_spectrum()
spectrumNoise.high_pass(6000)
filteredNoise = spectrumNoise.make_wave()
filteredNoise.make_audio()
```

```python
# after high_pass(6000)
spectrumNoise.plot(high= 10000)
```
